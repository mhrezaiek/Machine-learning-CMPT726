```python
import keras
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Add, concatenate
from tensorflow.keras.models import Model
import pandas as pd
import numpy as np 
from random import random
import time
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
import keras.backend as K
import pickle
from keras.initializers import RandomUniform, TruncatedNormal
```

```python
with open("/Users/parmis/Desktop/Machine-learning-CMPT726/datasets/cleaned/matrices/concatenated_matrix_with_passive.pickle", "rb") as pickle_file:
    content1 = pickle.load(pickle_file)
```

```python
with open("/Users/parmis/Desktop/Machine-learning-CMPT726/datasets/cleaned/matrices/features.pickle", "rb") as pickle_file:
    content2 = pickle.load(pickle_file)
```

```python
x = content1[:,:-1]
y = content1[:,-1]
y = np.asanyarray(y)
x =  np.asanyarray(x)
x1 = content2
x1 = np.asanyarray(x1)
#X_train, X_test, y_train, y_test= train_test_split(x, y, test_size=0.4, random_state=23)
```

```python
y = np.reshape(y, (17000,1))
```

```python
X = np.concatenate((x,x1),axis= 1) 
XY = np.concatenate((X,y), axis = 1)
np.shape(X)
```

```python
XY = XY.astype(float)
```

```python
XY = XY[~np.isnan(XY).any(axis=1)]
np.shape(XY)
```

```python
XY
```

```python
import plotly.express as px
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
components = pca.fit_transform(XY[:,:-1])
```

```python
import pylab as pl
for i in range(0, components.shape[0]):
    if XY[:,-1][i] == 0:
        c1 = pl.scatter(components[i,0],components[i,1],c='r',marker='+')
    elif XY[:,-1][i] == 1:
        c2 = pl.scatter(components[i,0],components[i,1],c='g',marker='o')
pl.legend([c1, c2], ['active', 'passive'])
pl.title('starbucks dataset with 2 clusters')
pl.show()
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
y_kmeans = kmeans.fit_predict(x)
```

```python
np.shape(y_kmeans)
```

```python
kmean=KMeans(n_clusters=2)
kmean.fit(x)
```

```python
plt.scatter(y[y_kmeans==0, 0], y[y_kmeans==0, 0], s=1000, c='red', label ='Cluster 1')
plt.scatter(y[y_kmeans==1, 0], y[y_kmeans==1, 0], s=1000, c='blue', label ='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
wcss = []
for i in range(1,20):
 kmeans = KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
 kmeans.fit(x)
 wcss.append(kmeans.inertia_)
 print("Cluster", i, "Inertia", kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title("The Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS") ##WCSS stands for total within-cluster sum of square
plt.show()
```
